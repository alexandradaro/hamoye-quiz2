```python
import pandas as pd
import matplotlib.pyplot  as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
dataset = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv")
dataset.head()
```

```python
dataset_new=dataset.drop(['date','lights'],axis=1)
dataset_new.head()
```

```python
scaler=MinMaxScaler()
normalised_dataset_new= pd.DataFrame(scaler.fit_transform(dataset_new), columns=dataset_new.columns)
```

```python
X=normalised_dataset_new.drop(['Appliances'], axis=1)
y=normalised_dataset_new['Appliances']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#question 12
traint2 = x_train[['T2']]
traint6 = x_train[['T6']]
testt2 = x_test[['T2']]
testt6 = x_test[['T6']]
```

```python
linearmodel = LinearRegression()
linearmodel.fit(traint2,traint6)
pred = linearmodel.predict(testt2)
```

```python
#Calculate the r2 score 
r2_value = round(r2_score(testt6,pred),2)
r2_value
```

```python
#question 13
#MAE
linearmodel
linearmodel.fit(x_train,y_train)
predicted_val=linearmodel.predict(x_test)

mae = round(mean_absolute_error(y_test,predicted_val),2)
mae
```

```python
#question14
#RSS
rss = np.sum(np.square(y_test - predicted_val))
round(rss, 2)
```

```python
#question15
#RMSE
rsme = np.sqrt(mean_squared_error(y_test, predicted_val))
round(rsme,3)
```

```python
#question16
#R2 score
r2_score =r2_score(y_test, predicted_val)
round(r2_score,2)
```

```python
#question17
def get_weights_dataset(linearmodel, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(linearmodel.coef_, feat.columns).sort_values()
  weights_dataset = pd.DataFrame(weights).reset_index()
  weights_dataset.columns = ['Features', col_name]
  weights_dataset[col_name].round(3)
  return weights_dataset

# Execute the get_weight function and store in dataframe
linear_weights_dataset = get_weights_dataset(linearmodel, x_train, 'Linear_weight')

linear_weights_dataset
```

```python
# question 18
#ridge regression
ridge_Reg = Ridge(alpha=0.04)
ridge_Reg.fit(x_train, y_train)
ridge_pred = ridge_Reg.predict(x_test)
ridge_rmse = np.sqrt(mean_squared_error(y_test, ridge_pred))
round(ridge_rmse, 3)
```

```python
#question 19
lasso = Lasso(alpha=0.001)
lasso.fit(x_train,y_train)
lasso_predicted_values= lasso.predict(x_test)

lasso_getweight_dataset =get_weights_dataset(lasso,x_train,'Lasso_weight')
non_zero_weight= lasso_getweight_dataset[lasso_getweight_dataset['Lasso_weight']!=0]

non_zero_weight
```

```python
#question20

#rsme with lasso
rsme_lasso =round(np.sqrt(mean_squared_error(y_test, lasso_predicted_values)), 3)
rsme_lasso
```
